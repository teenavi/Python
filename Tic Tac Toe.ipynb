```python


def board_array(board_arr,pos=0,token=0) :
    
    if(int(pos) >= 1 and int(pos) <= 9) :
        if(board_arr[pos] == " "):
            board_arr[pos] = token
        else:
            print("You already enter the value for the position "+pos)
            ask_position(board_arr,token,False)
        return True
    
    
def print_board(board_arr) :
    
    print(f" | "+board_arr["1"]+" | "+board_arr["2"]+" | "+board_arr["3"]+" | \n----------------\n | "+board_arr["4"]+" | "+board_arr["5"]+" | "+board_arr["6"]+" | \n---------------\n | "+board_arr["7"]+" | "+board_arr["8"]+" | "+board_arr["9"]+" |")
             
def alter_token(token,change_token):
    
    if(token == 'X' and change_token == True):
        token = '0'
    elif(token == '0' and change_token == True) :
        token = 'X'
    else:
        change_token = False
    return token


def check_result(board_arr,token):
    if (board_arr["1"] == board_arr["5"] == board_arr["9"] == 'X' or board_arr["3"] == board_arr["5"] == board_arr["7"] == 'X' or board_arr["2"] == board_arr["5"] == board_arr["8"] == 'X' or board_arr["4"] == board_arr["5"] == board_arr["6"] == 'X' or board_arr["3"] == board_arr["6"] == board_arr["9"] == 'X' or board_arr["1"] == board_arr["4"] == board_arr["7"] == 'X' or board_arr["1"] == board_arr["2"] == board_arr["3"] == 'X' or board_arr["7"] == board_arr["8"] == board_arr["9"] == 'X') :
        print("\nGame Over.\n")                
        print(" **** " +token + " won. ****")
        return True
    elif (board_arr["1"] == board_arr["5"] == board_arr["9"] == '0' or board_arr["3"] == board_arr["5"] == board_arr["7"] == '0' or board_arr["2"] == board_arr["5"] == board_arr["8"] == '0' or board_arr["4"] == board_arr["5"] == board_arr["6"] == '0' or board_arr["3"] == board_arr["6"] == board_arr["9"] == '0' or board_arr["1"] == board_arr["4"] == board_arr["7"] == '0' or board_arr["1"] == board_arr["2"] == board_arr["3"] == '0' or board_arr["7"] == board_arr["8"] == board_arr["9"] == '0') :
        print("\nGame Over.\n")                
        print(" **** " +token + " won. ****")
        return True
    else:
        dict_values = board_arr.values()
        value_count = list(dict_values).count('X')+list(dict_values).count('0')
        if(value_count == 9):
            print("\nGame Over.\n")                
            print(f"------------It's a Tie---------- !!")
            return True
        return False
    
        
def ask_position(board_arr,token,change_token) :
    
    print_board(board_arr)
    print("It's your turn,"+token+".Move to which place?")
    pos = input("")
    if(int(pos) > 9):
        print("You can enter only 1-9 positions")
        ask_position(board_arr,token,False)
    else :
        arr_status = board_array(board_arr,pos,token) #set the value to the dictionary
        if(arr_status) :
            rslt_status = check_result(board_arr,token) #check game status win/tie 
            if(rslt_status == False): #if not, show the game-board and continue the game
                token = alter_token(token,True)
                if(token):
                    ask_position(board_arr,token,True)
            else: #execute when the game win/tie
                board_arr[pos] = token #set the array with last entered value and display the board with the last entered token
                print_board(board_arr)
            
```
