```python
class calculator():
    
    #def __init__(self):
        #self.btn_obj = btn_obj
        
    def display_num(self,btn_obj) :
        
        self.btn_obj = btn_obj
        operator_list = ['+','-','*','/']
        self.count = 0
        self.text = self.btn_obj.cget("text")
        self.full_text = display.get()
        
        #display prevent from 03 to 3 
        if(self.full_text == '0'):
            display.delete(0,END)
            display.insert(0,'')
        
        #prevent display from +-(operators) together
        if(self.text == '+' or self.text == '-' or self.text == '*' or self.text == '/'):
            for i in operator_list :
                if i in self.full_text:
                    self.count += 1
            if self.count < 1 :       
                display.insert(100,self.text)
        else:
            display.insert(100,self.text)
            
        return self.text
            
    def display_clear(self):
        display.delete(0,END)
        
    def display_backspace(self):
        #take the string currently in the widget, all the way up to the last character
        self.txt = display.get()[:-1]
        #clear the widget of text
        display.delete(0, tkinter.END)
        #insert the new string, sans the last character
        display.insert(0, self.txt)
        
    def clear_list(self):
        operand_list = []
        
    def display_result(self):
        
        self.txt = display.get()
        operator_list = ['+','-','*','/']
        for i in operator_list :
            if i in self.txt:
                operand_list = self.txt.split(i)
                
        if '+' in self.txt:
            if operand_list[0].isdecimal() and operand_list[1].isdecimal():
                result = int(operand_list[0])+int(operand_list[1])
            elif operand_list[0].isdecimal() or operand_list[1].isdecimal():
                if operand_list[0].isdecimal():
                    result = int(operand_list[0])+float(operand_list[1])
                else:
                    result = float(operand_list[0])+int(operand_list[1])
            else:
                result = float(operand_list[0])+float(operand_list[1])
            self.clear_list()
            return result
        elif '-' in self.txt:
            
            if operand_list[0].isdecimal() and operand_list[1].isdecimal():
                result = int(operand_list[0])-int(operand_list[1])
            elif operand_list[0].isdecimal() or operand_list[1].isdecimal():
                if operand_list[0].isdecimal():
                    result = int(operand_list[0])-float(operand_list[1])
                else:
                    result = float(operand_list[0])-int(operand_list[1])
            else:
                result = float(operand_list[0])-float(operand_list[1])
            self.clear_list()
            return result
        
        elif '*' in self.txt:
            
            if operand_list[0].isdecimal() and operand_list[1].isdecimal():
                result = int(operand_list[0])*int(operand_list[1])
            elif operand_list[0].isdecimal() or operand_list[1].isdecimal():
                if operand_list[0].isdecimal():
                    result = int(operand_list[0])*float(operand_list[1])
                else:
                    result = float(operand_list[0])*int(operand_list[1])
            else:
                result = float(operand_list[0])*+float(operand_list[1])
            self.clear_list()
            return result
        
        elif '/' in self.txt:
            
            if operand_list[0].isdecimal() and operand_list[1].isdecimal():
                result = int(operand_list[0])/int(operand_list[1])
            elif operand_list[0].isdecimal() or operand_list[1].isdecimal():
                if operand_list[0].isdecimal():
                    result = int(operand_list[0])/float(operand_list[1])
                else:
                    result = float(operand_list[0])/int(operand_list[1])
            else:
                result = float(operand_list[0])/float(operand_list[1])
            self.clear_list()
            return result
        
    def display_oprns(self,btn_obj):
        
        self.btn_obj = btn_obj
        flag = False
        self.btntext = self.btn_obj.cget("text")
        self.txt = display.get()
        operator_list = ['+','-','*','/']
        for i in operator_list :
            if i in self.txt:
                flag = True
        if self.txt == '':
            result = '0'
        else:
            if self.txt.isdecimal()  and flag == False:
                if self.btntext == 'sqrt':
                    result = math.sqrt(int(self.txt))
                elif self.btntext == '^2':
                    result = int(self.txt)**2
                elif self.btntext == '1/x':
                    result = 1/int(self.txt)
            elif self.txt.isdecimal() == False and flag == False:
                if self.btntext == 'sqrt':
                    result = math.sqrt(float(self.txt))
                elif self.btntext == '^2':
                    result = float(self.txt)**2
                elif self.btntext == '1/x':
                    result = 1/float(self.txt)
            else:
                result = '0'
        return result
    
    def percentage(self):
        
        self.txt = display.get()
        if '/' in self.txt:
            operand_list = self.txt.split('/')
            if operand_list[0].isdecimal() and operand_list[1].isdecimal():
                result = int(operand_list[0])/int(operand_list[1])*100
            elif operand_list[0].isdecimal() or operand_list[1].isdecimal():
                if operand_list[0].isdecimal():
                    result = int(operand_list[0])/float(operand_list[1])*100
                else:
                    result = float(operand_list[0])/int(operand_list[1])*100
            else:
                result = float(operand_list[0])/float(operand_list[1])*100
                self.clear_list()
            return result
        else :
            return '0'
```

```python
calc = calculator()
operand_list =[]

def set_num(btn_obj):
    text = calc.display_num(btn_obj)
    
def clear_all():
    calc.display_clear()
    
def backspace():
    calc.display_backspace()
        
def result():
    res = calc.display_result()
    if(res):
        display.delete(0,END)
        display.insert(100,res)
        operand_list.append(res)

def operations(btn_obj):
    res = calc.display_oprns(btn_obj)
    if(res):
        display.delete(0,END)
        display.insert(100,res)
        operand_list.append(res)
        
def percentage_oprn():
    res = calc.percentage()
    if(res):
        display.delete(0,END)
        display.insert(100,res)
        operand_list.append(res)
```
